```python
obj_col = [i for i in train.columns if train[i].dtypes != 'int64']
int_col = [i for i in train.columns if train[i].dtypes == 'int64']
```

```python
# 범주형 변수 

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(8,18))

for i, col in enumerate(obj_col[1:]):
    plt.subplot(8, 2, i+1)
    plt.title(col)
    sns.countplot(data=train, x=train[col])
    
plt.tight_layout()
plt.show();
```

```python
sns.pairplot(train[int_col])
```

```python
import copy
cp_df = copy.deepcopy(train)

# # One-Hot
# for col in obj_col:
#     dummy = pd.get_dummies(train[col])
#     cp_df = pd.concat([cp_df, dummy], axis=1)
    
# Label
for col in obj_col[1:]:
    cp_df[col+'_Label'] = cp_df[col].map({j:i for i, j in enumerate(train[col].unique())})
    
cp_df
```

```python
# Label Encoding 된 데이터로 상관관계 분석
drop_train = cp_df.drop(columns=obj_col)
drop_train
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(drop_train.corr(), annot=True, cmap='Blues')
plt.show();
```

```python
# 상관관계 조건에 해당하는 컬럼만 추출

corr_df = drop_train.corr()

corr_cond = corr_df[(corr_df >= 0.1) | (corr_df <= - 0.1)].loc['Income']

corr_col = [i for i, data in zip(corr_cond.index, corr_cond) if not np.isnan(data)]
corr_col
```

```python
# 테스트 데이터 전처리

import copy
cp_test = copy.deepcopy(test.fillna(0))

# # One-Hot
# for col in obj_col:
#     dummy = pd.get_dummies(train[col])
#     cp_df = pd.concat([cp_df, dummy], axis=1)
    
# Label
for col in obj_col[1:]:
    cp_test[col+'_Label'] = cp_test[col].map({j:i for i, j in enumerate(train[col].unique())})
    
cp_test
```

```python
# Label Encoding 된 데이터로 상관관계 분석
drop_test = cp_test.drop(columns=obj_col)
drop_test
```

```python
# 종속변수 분리

y_train = cp_df[corr_col].Income
x_train = cp_df[corr_col].drop(columns=['Income'])
```

```python
x_test = drop_test[x_train.columns]
x_test
```

```python
x_train
```

```python
from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, random_state=2024, test_size=0.2)
```

```python
# Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_train = scaler.fit_transform(x_tr)
scaled_val = scaler.transform(x_val)
```

```python
scaled_x = scaler.fit_transform(x_train)
scaled_test = scaler.transform(x_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

def dnn(num_input=1):
    model = Sequential()
    model.add(Dense(512, activation='relu', input_dim=num_input))
    
    model.add(BatchNormalization())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.3))
    
    model.add(Dense(128, activation='relu'))
    
    model.add(BatchNormalization())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    
    model.add(Dense(32, activation='relu'))
    
    model.add(BatchNormalization())
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.3))
    
    model.add(Dense(1, activation='linear'))
    
    model.compile(optimizer='adam', loss='mse', metrics=['mse'])
    
    return model

model = dnn(num_input=9)

# 미니 배치 학습
model.fit(scaled_train, y_tr, epochs=100, batch_size=32, verbose=2)

# 모델 평가
model.evaluate(scaled_val, y_val)

# 교차 검증
model = dnn(num_input=9)
history = model.fit(scaled_x, y_train, batch_size=32, epochs=100, validation_split=0.2, verbose=2)

# 
import matplotlib.pyplot as plt

def plot_loss_curve(total_epoch=10, start=1):
    plt.figure(figsize=(15, 5))
    plt.plot(range(start, total_epoch +1), history.history['loss'][start-1:total_epoch], label='Train')
    plt.plot(range(start, total_epoch +1), history.history['val_loss'][start-1:total_epoch], label='Validation')
    plt.xlabel('Epochs')
    plt.ylabel('mse')
    plt.legend()
    plt.show()
    
plot_loss_curve(total_epoch=100, start=1)

plot_loss_curve(total_epoch=100, start=20)
```

```python
pred = model.predict(x_test)
```

```python
result_tf = pd.concat([test.ID, pd.DataFrame(pred, columns=['Income'])], axis=1).fillna(0)
result_tf
```
